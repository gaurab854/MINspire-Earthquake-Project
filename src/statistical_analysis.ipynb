# Statistical Analysis
Phase 8: Decadal Statistics, Correlation Matrix

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set_theme(style='whitegrid')

BASE_DIR = os.path.abspath('..')
OUTPUT_DIR = os.path.join(BASE_DIR, 'outputs')
df = pd.read_csv(os.path.join(OUTPUT_DIR, 'cleaned_earthquake_data.csv'), parse_dates=['Datetime'])
print(f'Loaded {len(df)} records.')
```

## Statistics by Decade

```python
decade_stats = df.groupby('Decade').agg(
    Events=('Magnitude','count'),
    Avg_Magnitude=('Magnitude','mean'),
    Max_Magnitude=('Magnitude','max'),
    Avg_Depth=('Depth','mean')
).round(2)
print('=== Decadal Statistics ===')
display(decade_stats)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
decade_stats['Events'].plot(kind='bar', ax=axes[0], color='steelblue', edgecolor='k')
axes[0].set_title('Earthquake Events per Decade'); axes[0].set_xlabel('Decade'); axes[0].set_ylabel('Count')
decade_stats['Avg_Magnitude'].plot(kind='line', ax=axes[1], marker='o', color='crimson')
axes[1].set_title('Avg Magnitude per Decade'); axes[1].set_xlabel('Decade'); axes[1].set_ylabel('Avg Magnitude')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'decadal_stats.png'), dpi=150)
plt.show()
```

## Annual Earthquake Frequency

```python
annual_freq = df['Year'].value_counts().sort_index()
print(f'Average earthquakes per year: {annual_freq.mean():.2f}')
print(f'Peak year: {annual_freq.idxmax()} ({annual_freq.max()} events)')
```

## Correlation Matrix

```python
corr_cols = [c for c in ['Magnitude','Depth','Latitude','Longitude','Year'] if c in df.columns]
corr = df[corr_cols].corr().round(3)
display(corr)

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix', fontsize=13)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'correlation_matrix.png'), dpi=150)
plt.show()
```

## Save Statistical Summary

```python
with open(os.path.join(OUTPUT_DIR, 'statistical_summary.txt'), 'w', encoding='utf-8') as f:
    f.write('=== Statistical Analysis Summary ===\n\n')
    f.write('1. Statistics by Decade:\n')
    f.write(decade_stats.to_string() + '\n\n')
    f.write(f'2. Average Earthquakes per Year: {annual_freq.mean():.2f}\n\n')
    f.write('3. Correlation Matrix:\n')
    f.write(corr.to_string() + '\n')
print('Statistical summary saved.')
```
