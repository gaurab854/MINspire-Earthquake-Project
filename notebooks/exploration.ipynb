# Seismic Hotspot Evolution in Nepal (1990-2026)
This notebook demonstrates the interactive exploration and clustering of seismic hotspots in Nepal.

```python
%matplotlib inline
import os
import sys

# Ensure project root is in path
sys.path.append(os.path.abspath('..'))

import pandas as pd
import matplotlib.pyplot as plt
from src.config import DATA_DIR, OUTPUT_DIR
from src.data_processing import get_data_path, clean_data, engineer_features
from src.eda import perform_eda
from src.clustering import detect_hotspots, analyze_hotspot_evolution
from src.statistical_analysis import perform_statistical_analysis
from src.risk_analysis import identify_risk_zones
```

## 1. Load and Clean Dataset

```python
data_path = get_data_path(DATA_DIR)
print(f"Using dataset: {data_path}")
df = pd.read_csv(data_path)
df_clean = clean_data(df)
df_eng = engineer_features(df_clean)
df_eng.head()
```

## 2. Interactive Data Visualizations

```python
perform_eda(df_eng, OUTPUT_DIR)
plt.figure(figsize=(10, 6))
plt.scatter(df_eng['Longitude'], df_eng['Latitude'], c=df_eng['Magnitude'], cmap='flare', alpha=0.5)
plt.colorbar(label='Magnitude')
plt.title('Earthquake Epicenters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

## 3. Hotspot Clustering

```python
df_clustered = detect_hotspots(df_eng, OUTPUT_DIR)
df_evolved = analyze_hotspot_evolution(df_clustered, OUTPUT_DIR)
```

## 4. Statistics and Risk Analysis

```python
perform_statistical_analysis(df_evolved, OUTPUT_DIR)
identify_risk_zones(df_evolved, OUTPUT_DIR)
```
